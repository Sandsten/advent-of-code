```python
tiles = open("input/input-day-20.txt", "r").read().split("\n")
tiles.insert(0,"") # To make it uniform

tiles_dictionary = {}

for i in range(0,len(tiles)-12,12):
    # The first rown is the id and the id is always at the end and 4 numbers
    tile_id = tiles[i+1][-5:-1]
    # The remaining rows make up the tile
    tile = tiles[i+2:i+12] 
    # Make each "camera" or "pixel" its own element in a multidim array
    tile_matrix = [[x for x in t] for t in tile]
    # Store them in a dictionary
    tiles_dictionary[tile_id] = tile_matrix
```
