### Dijkstra's algorithm  



```python
import math

width = len(data[0])
height = len(data)

start_pos = None
end_pos = None
for y in range(height):
  for x in range(width):
    if data[y][x] == "S":
      start_pos = [x,y]
    elif data[y][x] == "E":
      end_pos = [x,y]

print(f"Map stats:\n\twidth = {width}\n\theight = {height}")
print(f"Start pos = {start_pos}")
print(f"End pos = {end_pos}")

def getHeight(x):
  letters = "abcdefghijklmnopqrstuvwxyz"
  if x == "S":
    return letters.index("a")
  if x == "E":
    return letters.index("z")
  return letters.index(x)

unvisited_set = [[True for x in row] for row in data]

tentative_distance = [[math.inf for x in row] for row in data]
tentative_distance[start_pos[1]][start_pos[0]] = 0

previous_vertex = [[None for x in row] for row in data]

Q = [start_pos]

while Q != []:
  # find vertex in Q with the shortest tentative distance. That's the next vertex to check
  vertex_with_shortest = 0
  shortest_distance = math.inf
  for i in range(len(Q)):
    vx, vy = Q[i]
    distance = tentative_distance[vy][vx]
    if distance < shortest_distance:
      vertex_with_shortest = [vx,vy]
      shortest_distance = distance
  
  v = Q.pop(i)

  # Find all unvisited neighbors to vertex
  unvisited_neighbors = []
  for n_dir in [[1,0],[-1,0],[0,1],[0,-1]]:
    nx = v[0] + n_dir[0]
    ny = v[1] + n_dir[1]

    if nx < 0 or ny < 0 or nx >= width or ny >= height:
      continue
    
    # check if neighbor has maximum height diff of + 1
    h_current = getHeight(data[v[1]][v[0]])
    h_neighbor = getHeight(data[ny][nx])
    if h_neighbor > h_current + 1:
      continue

    if unvisited_set[ny][nx]:
        unvisited_neighbors.append([nx, ny])

  # Check if distance is shorter
  for neighbor in unvisited_neighbors:
    n_x, n_y = neighbor 
    alt = tentative_distance[v[1]][v[0]] + 1 # All edges have the same length
    if alt < tentative_distance[n_y][n_x]:
      tentative_distance[n_y][n_x] = alt
      previous_vertex[n_y][n_x] = v
      Q.append(neighbor)
  
  unvisited_set[v[1]][v[0]] = False


previous = []
for row in previous_vertex:
  for value in row:
    if value != None:
      previous.append(value)


print(f"Shortest")


path = []
previous = previous_vertex[end_pos[1]][end_pos[0]]
while previous != None:
  px,py = previous
  path.append([px,py])
  previous = previous_vertex[py][px]

for y in range(height):
  for x in range(width):
    if [x,y] in path:
      print(" ", end="")
    else:
      print(data[y][x], end="")
  print()
```
