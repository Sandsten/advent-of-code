```python
with open("input/day12.txt", "r") as f:
  data = f.read().splitlines()
```

### Dijkstra's algorithm  



```python
import math

width = len(data[0])
height = len(data)

start_pos = None
end_pos = None
for y in range(height):
  for x in range(width):
    if data[y][x] == "S":
      start_pos = [x,y]
    elif data[y][x] == "E":
      end_pos = [x,y]

def getHeight(x):
  letters = "abcdefghijklmnopqrstuvwxyz"
  if x == "S":
    return letters.index("a")
  if x == "E":
    return letters.index("z")
  return letters.index(x)

unvisited_set = [[True for x in row] for row in data]

tentative_distance = [[math.inf for x in row] for row in data]
tentative_distance[start_pos[1]][start_pos[0]] = 0

previous_vertex = [[None for x in row] for row in data]

Q = [start_pos]

while Q != []:
  vx,vy = Q.pop()

  # Find all unvisited neighbors to vertex
  unvisited_neighbors = []
  for n_dir in [[1,0],[-1,0],[0,-1],[0,1]]:
    nx = vx + n_dir[0]
    ny = vy + n_dir[1]

    if nx < 0 or ny < 0 or nx >= width or ny >= height:
      continue
    
    # check if neighbor has maximum height diff of + 1
    h_current = getHeight(data[vy][vx])
    h_neighbor = getHeight(data[ny][nx])
    if h_neighbor > h_current + 1:
      continue

    if unvisited_set[ny][nx]:
        unvisited_neighbors.append([nx, ny])

  # Check if distance is shorter
  for neighbor in unvisited_neighbors:
    nx, ny = neighbor 
    alt = tentative_distance[vy][vx] + 1 # All edges have the same length
    if alt < tentative_distance[ny][nx]:
      tentative_distance[ny][nx] = alt
      previous_vertex[ny][nx] = [vx, vy]
      Q.insert(0,neighbor)
  
  unvisited_set[vy][vx] = False

path = []
previous = previous_vertex[end_pos[1]][end_pos[0]]
while previous != None:
  px,py = previous
  path.append([px,py])
  previous = previous_vertex[py][px]

print(f"Fewest nr of steps to get to goal is {len(path)}")

for y in range(height):
  for x in range(width):
    if [x,y] in path:
      print(" ", end="")
    else:
      print(data[y][x], end="")
  print()
```

```python
def findShortestPath(start) -> int:
  start_pos = start

  unvisited_set = [[True for x in row] for row in data]

  tentative_distance = [[math.inf for x in row] for row in data]
  tentative_distance[start_pos[1]][start_pos[0]] = 0

  previous_vertex = [[None for x in row] for row in data]

  Q = [start_pos]

  while Q != []:
    vx,vy = Q.pop()

    # Find all unvisited neighbors to vertex
    unvisited_neighbors = []
    for n_dir in [[1,0],[-1,0],[0,-1],[0,1]]:
      nx = vx + n_dir[0]
      ny = vy + n_dir[1]

      if nx < 0 or ny < 0 or nx >= width or ny >= height:
        continue
      
      # check if neighbor has maximum height diff of + 1
      h_current = getHeight(data[vy][vx])
      h_neighbor = getHeight(data[ny][nx])
      if h_neighbor > h_current + 1:
        continue

      if unvisited_set[ny][nx]:
          unvisited_neighbors.append([nx, ny])

    # Check if distance is shorter
    for neighbor in unvisited_neighbors:
      nx, ny = neighbor 
      alt = tentative_distance[vy][vx] + 1 # All edges have the same length
      if alt < tentative_distance[ny][nx]:
        tentative_distance[ny][nx] = alt
        previous_vertex[ny][nx] = [vx, vy]
        Q.insert(0,neighbor)
    
    unvisited_set[vy][vx] = False

  path = []
  previous = previous_vertex[end_pos[1]][end_pos[0]]
  while previous != None:
    px,py = previous
    path.append([px,py])
    previous = previous_vertex[py][px]

  return len(path)

all_a_vertices = []

for y in range(height):
  for x in range(width):
    if data[y][x] == "a" or data[y] == "S":
      all_a_vertices.append([x, y])

steps = []
for start in all_a_vertices:
  nr_steps = findShortestPath(start)
  # Some starting positions will get you stuck. I.e you are in a valley and can't get out
  if nr_steps > 0:
    steps.append(nr_steps)

steps.sort()

print(f"Shortest path to E from any a start pos = {steps[0]}")
```
